```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import re
```

```python
lang = pd.read_csv("bar_data.csv")
```

```python
lang.set_index("Responder_id", inplace=True)
```

```python
first_list = lang["LanguagesWorkedWith"].tolist()

lang_list = list()

for i in first_list:
    lang_list +=  i.split(";")

lang_dict = dict()

for i in lang_list:
    lang_dict[i] = lang_dict.get(i, 0) + 1
```

```python
df = pd.DataFrame((lang_dict.items()), columns=["Languages","Numbers"])
df.set_index("Languages", inplace=True)
df.sort_values("Numbers", inplace=True)
df
```

```python
ax = df.plot.barh(figsize=(10,10))
ax.axvline(30000)
plt.tight_layout()
```
